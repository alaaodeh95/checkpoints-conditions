```python
import pandas as pd
import numpy as np
from pymongo import MongoClient
from datetime import datetime
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
client = MongoClient('mongodb://localhost:27017/')

db = client['RoadsConditions']
collection = db['StructuredData']

data = list(collection.find())
df = pd.DataFrame(data)

if '_id' in df.columns:
    df.drop('_id', axis=1, inplace=True)

df = df[df['checkpoint'].isin(["المربعة","عورتا","دير شرف"])]
df = df[df['status'] == 'مغلق']

df.head()
```

```python
status_mapping = {
    "مفتوح": "Open",
    "حاجز": "Open",
    "ازمة خفيفة": "Traffic",
    "ازمة خانقة": "Traffic",
    "ازمة": "Traffic"
}

df['status_mapped'] = df['status'].map(status_mapping)
unmapped_status = df[df['status_mapped'].isnull()]['status'].unique()
if len(unmapped_status) > 0:
    print("Unmapped Statuses found:", unmapped_status)
```

```python
# Define the mapping for direction
direction_mapping = {
    "داخل": "Enter",
    "خارج": "Exit",
    "الاتجاهين": "Both"
}

# Apply the mapping to create 'direction_mapped' column
df['direction_mapped'] = df['direction'].map(direction_mapping)

# Check for any unmapped directions
unmapped_directions = df[df['direction_mapped'].isnull()]['direction'].unique()
if len(unmapped_directions) > 0:
    print("Unmapped Directions found:", unmapped_directions)
    # Handle unmapped directions as needed. For example, drop them:
    df = df[df['direction_mapped'].notnull()].copy()
    print("Dropped records with unmapped directions.")
else:
    print("All directions mapped successfully.")
```

```python
# Separate records based on 'direction_mapped'
df_enter = df[df['direction_mapped'] == 'Enter'].copy()
df_exit = df[df['direction_mapped'] == 'Exit'].copy()
df_both = df[df['direction_mapped'] == 'Both'].copy()


df_both_enter = df_both.copy()
df_both_enter['direction_mapped'] = 'Enter'

df_both_exit = df_both.copy()
df_both_exit['direction_mapped'] = 'Exit'

df = pd.concat([df_enter, df_exit, df_both_enter, df_both_exit], ignore_index=True)
```

```python
# Extract hour (0-23)
df['hour'] = df['sentTime'].dt.hour

# Extract day of the week (Monday=0, Sunday=6)
df['day_of_week'] = df['sentTime'].dt.dayofweek

# Optionally, map day numbers to names
day_mapping = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}

df['day_name'] = df['day_of_week'].map(day_mapping)

# Drop the original 'sentTime' as it's no longer needed
df.drop(['sentTime'], axis=1, inplace=True)

# Display the updated DataFrame
df.head()
```

```python
# Cyclical encoding for hour
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

# Cyclical encoding for day of the week
df['day_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['day_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)

# One-Hot Encode 'checkpoint'
checkpoint_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
checkpoint_encoded = checkpoint_encoder.fit_transform(df[['checkpoint']])
checkpoint_df = pd.DataFrame(checkpoint_encoded, columns=checkpoint_encoder.get_feature_names_out(['checkpoint']))

# Concatenate the encoded checkpoint data
df = pd.concat([df, checkpoint_df], axis=1)

# Drop the original 'checkpoint' and 'day_name' columns
df.drop(['checkpoint', 'day_of_week', 'day_name', 'hour'], axis=1, inplace=True)

# Display the updated DataFrame
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder with updated parameter to avoid FutureWarning
direction_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')  # Use 'sparse_output' instead of 'sparse'

# Fit and transform the 'direction_mapped' column
direction_encoded = direction_encoder.fit_transform(df[['direction_mapped']])

# Create a DataFrame from the encoded features
direction_df = pd.DataFrame(direction_encoded, columns=direction_encoder.get_feature_names_out(['direction_mapped']))

# Concatenate the encoded direction data with the original DataFrame
df = pd.concat([df, direction_df], axis=1)

# Drop the original 'direction' and 'direction_mapped' columns if no longer needed
df.drop(['direction', 'direction_mapped'], axis=1, inplace=True)

print("\nDataFrame after encoding 'direction_mapped':")
display(df.head())
```

```python
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf

# Initialize LabelEncoder
le_status = LabelEncoder()

# Encode the 'status_mapped' labels
y_encoded = le_status.fit_transform(df['status_mapped'])

# One-Hot Encode the labels
y_ohe = tf.keras.utils.to_categorical(y_encoded, num_classes=3)

# Add the encoded labels to the DataFrame (optional)
df['status_encoded'] = y_encoded
df['status_ohe_0'] = y_ohe[:, 0]
df['status_ohe_1'] = y_ohe[:, 1]
df['status_ohe_2'] = y_ohe[:, 2]

print("\nSample of encoded status:")
display(df[['status', 'status_mapped', 'status_encoded', 'status_ohe_0', 'status_ohe_1', 'status_ohe_2']].head())
```

```python
# Get the names of the one-hot encoded direction columns
direction_feature_names = direction_encoder.get_feature_names_out(['direction_mapped'])

# Define feature columns including direction features
feature_columns = [
    'hour_sin',
    'hour_cos',
    'day_sin',
    'day_cos'
] + list(checkpoint_df.columns) + list(direction_feature_names)

X = df[feature_columns].values

# Select target
y = y_ohe  # This is the one-hot encoded status

# Verify the shapes
print("\nShape of feature matrix X:", X.shape)
print("Shape of target matrix y:", y.shape)
```

```python
y_labels = np.argmax(y_ohe, axis=1)

# Display the unique classes and their corresponding integer labels
unique_classes = le_status.classes_
print("Class Labels Mapping:")
for idx, class_name in enumerate(unique_classes):
    print(f"{idx}: {class_name}")
    
from sklearn.model_selection import train_test_split

# Split the data: 70% train, 30% temp
X_train, X_temp, y_train_ohe, y_temp_ohe, y_train_labels, y_temp_labels = train_test_split(
    X, y_ohe, y_labels, test_size=0.3, random_state=42, stratify=y_labels
)

# Split temp into 50% validation and 50% test (each 15% of total)
X_val, X_test, y_val_ohe, y_test_ohe, y_val_labels, y_test_labels = train_test_split(
    X_temp, y_temp_ohe, y_temp_labels, test_size=0.5, random_state=42, stratify=y_temp_labels
)

# Verify the splits
print(f'Training samples: {X_train.shape[0]}')
print(f'Validation samples: {X_val.shape[0]}')
print(f'Test samples: {X_test.shape[0]}')
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# Initialize individual classifiers
clf1 = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs')
clf2 = DecisionTreeClassifier(random_state=42)
clf3 = SVC(probability=True, random_state=42)

# Initialize Voting Classifier
voting_clf = VotingClassifier(estimators=[
    ('lr', clf1), 
    ('dt', clf2), 
    ('svc', clf3)
], voting='soft')

# Train the ensemble
voting_clf.fit(X_train, y_train_labels)

# Predict on test set
y_pred = voting_clf.predict(X_test)

# Evaluate
accuracy = accuracy_score(y_test_labels, y_pred)
print(f"Voting Classifier Test Accuracy: {accuracy:.4f}")

# Classification Report
print("\nVoting Classifier Classification Report:")
print(classification_report(y_test_labels, y_pred, target_names=le_status.classes_))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the classifier
rf_clf = RandomForestClassifier(n_estimators=300, random_state=42)

# Train the classifier
rf_clf.fit(X_train, y_train_labels)

# Predict on test set
y_pred_rf = rf_clf.predict(X_test)

# Evaluate
accuracy_rf = accuracy_score(y_test_labels, y_pred_rf)
print(f"Random Forest Test Accuracy: {accuracy_rf:.4f}")

# Classification Report
print("\nRandom Forest Classification Report:")
print(classification_report(y_test_labels, y_pred_rf, target_names=le_status.classes_))
```

```python
# Example: Re-defining the model (if needed)
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout

# Define the input shape based on the number of features
input_dim = X_train.shape[1]

# Input Layer
inputs = Input(shape=(input_dim,))

# Hidden Layers
x = Dense(128, activation='relu')(inputs)
x = Dropout(0.5)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.5)(x)

# Output Layer
output = Dense(y_train_ohe.shape[1], activation='softmax', name='status_output')(x)

# Define the Model
model = Model(inputs=inputs, outputs=output)

# Compile the Model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Display the Model Summary
model.summary()

# Define Early Stopping Callback
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

# Train the Model
history = model.fit(
    X_train,
    y_train_ohe,
    epochs=100,
    batch_size=32,
    validation_data=(X_val, y_val_ohe),
    callbacks=[early_stopping]
)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

clf1 = LogisticRegression(max_iter=1000, random_state=42)
clf2 = RandomForestClassifier(n_estimators=100, random_state=42)
clf3 = SVC(probability=True, random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', clf1), ('rf', clf2), ('svc', clf3)],
    voting='soft'
)

voting_clf.fit(X_train, y_train_labels)
y_pred_voting = voting_clf.predict(X_test)
accuracy_voting = accuracy_score(y_test_labels, y_pred_voting)
print(f"Voting Classifier Test Accuracy: {accuracy_voting:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, 
                           cv=5, n_jobs=-1, scoring='accuracy')
grid_search.fit(X_train, y_train_labels)

print(f"Best Parameters: {grid_search.best_params_}")
best_rf = grid_search.best_estimator_
y_pred_best_rf = best_rf.predict(X_test)
accuracy_best_rf = accuracy_score(y_test_labels, y_pred_best_rf)
print(f"Best Random Forest Test Accuracy: {accuracy_best_rf:.4f}")
```

```python
from sklearn.utils import class_weight
from imblearn.over_sampling import SMOTE
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

# Calculate class weights
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_labels),
    y=y_train_labels
)

class_weights_dict = dict(enumerate(class_weights))
print("Class Weights:", class_weights_dict)

# 2. Handling Class Imbalance with SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to the training data
X_train_smote, y_train_labels_smote = smote.fit_resample(X_train, y_train_labels)

# Convert resampled labels back to one-hot encoding for Keras models
y_train_ohe_smote = tf.keras.utils.to_categorical(y_train_labels_smote, num_classes=len(unique_classes))

# Verify new class distribution after SMOTE
class_counts_smote = pd.Series(y_train_labels_smote).value_counts().sort_index()
class_df_smote = pd.DataFrame({'Class': unique_classes, 'Count': class_counts_smote})
plt.figure(figsize=(8,6))
sns.barplot(x='Class', y='Count', data=class_df_smote, palette='magma')
plt.title('Resampled Class Distribution with SMOTE')
plt.xlabel('Status')
plt.ylabel('Number of Samples')
plt.show()

# 3. Define the Neural Network Model

def create_model(input_dim, num_classes):
    inputs = Input(shape=(input_dim,))
    x = Dense(128, activation='relu')(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(64, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    model = Model(inputs=inputs, outputs=outputs)
    return model

# Instantiate the model
model = create_model(input_dim=X_train.shape[1], num_classes=y_train_ohe.shape[1])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Display the Model Summary
model.summary()

# Define Early Stopping Callback
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

# 4. Train the Model with Class Weights

print("\nTraining with Class Weights:")
history_class_weights = model.fit(
    X_train,
    y_train_ohe,
    epochs=100,
    batch_size=32,
    validation_data=(X_val, y_val_ohe),
    callbacks=[early_stopping],
    class_weight=class_weights_dict
)

# 5. Train the Model with SMOTE

# Re-instantiate the model to train from scratch
model_smote = create_model(input_dim=X_train.shape[1], num_classes=y_train_ohe.shape[1])
model_smote.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])

print("\nTraining with SMOTE:")
history_smote = model_smote.fit(
    X_train_smote,
    y_train_ohe_smote,
    epochs=100,
    batch_size=32,
    validation_data=(X_val, y_val_ohe),
    callbacks=[early_stopping]
)

# 6. Evaluate Both Models on the Test Set

# Function to Evaluate and Plot Results
def evaluate_model(model, X_test, y_test_ohe, y_test_labels, model_name):
    print(f"\nEvaluating {model_name} on Test Set:")
    test_loss, test_accuracy = model.evaluate(X_test, y_test_ohe, verbose=0)
    print(f"Test Accuracy: {test_accuracy:.4f}")

    # Predict Class Probabilities
    y_pred_prob = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred_prob, axis=1)
    y_true_classes = y_test_labels

    # Classification Report
    print(f"\nClassification Report for {model_name}:")
    print(classification_report(y_true_classes, y_pred_classes, target_names=unique_classes))

    # Confusion Matrix
    cm = confusion_matrix(y_true_classes, y_pred_classes)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=unique_classes,
                yticklabels=unique_classes)
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Evaluate Model Trained with Class Weights
evaluate_model(model, X_test, y_test_ohe, y_test_labels, "Neural Network with Class Weights")

# Evaluate Model Trained with SMOTE
evaluate_model(model_smote, X_test, y_test_ohe, y_test_labels, "Neural Network with SMOTE")

# 7. (Optional) Evaluate Other Models (e.g., Random Forest)

from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest with Class Weights
rf_clf = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',
    random_state=42
)

# Train the Random Forest Classifier
rf_clf.fit(X_train, y_train_labels)

# Predict on Test Set
y_pred_rf = rf_clf.predict(X_test)

# Evaluate Random Forest
print("\nEvaluating Random Forest on Test Set:")
accuracy_rf = rf_clf.score(X_test, y_test_labels)
print(f"Random Forest Test Accuracy: {accuracy_rf:.4f}")

# Classification Report for Random Forest
print("\nClassification Report for Random Forest:")
print(classification_report(y_test_labels, y_pred_rf, target_names=unique_classes))

# Confusion Matrix for Random Forest
cm_rf = confusion_matrix(y_test_labels, y_pred_rf)
plt.figure(figsize=(6,4))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens',
            xticklabels=unique_classes,
            yticklabels=unique_classes)
plt.title('Confusion Matrix for Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
